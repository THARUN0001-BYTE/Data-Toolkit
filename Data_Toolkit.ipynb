**1.  What is NumPy, and why is it widely used in Python?**

**Answer:**NumPy (Numerical Python) is a fundamental Python library for numerical and scientific computing. It provides powerful tools for working with large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.

**1. Efficient Array Operations:**

NumPy arrays (ndarray) are more compact and faster than Python’s built-in lists for numerical data.

Operations on arrays are performed element-wise, leveraging highly optimized C and Fortran code under the hood.

**2. Multidimensional Arrays:**

Supports n-dimensional arrays (1D, 2D, 3D, etc.), making it ideal for mathematical, image, and signal processing tasks.

**3. Broadcasting:**

Allows operations on arrays of different shapes and sizes without writing explicit loops.

**4. Mathematical Functions:**

Provides a vast set of mathematical functions such as linear algebra, Fourier transforms, statistics, and more.

**5. Integration with Other Libraries:**

Libraries like Pandas, SciPy, Scikit-learn, TensorFlow, and Matplotlib heavily rely on NumPy arrays.

**6. Memory Efficiency:**

NumPy uses less memory and is more efficient due to its fixed-type arrays (unlike Python lists, which can hold mixed types).

**7. Interfacing with C/C++:**

Allows easy integration with code written in C, C++, and Fortran for high-performance computing tasks.



```python
# Example:

import numpy as np

# Creating an array
a = np.array([1, 2, 3])

# Element-wise operations
b = a * 2   # [2, 4, 6]

# Mean of array
mean_val = np.mean(b)  # 4.0
```

**2.  How does broadcasting work in NumPy?**

**Answer:** Broadcasting in NumPy allows you to perform arithmetic operations on arrays of different shapes without explicitly resizing them. It automatically expands the smaller array so that it matches the shape of the larger one without actually copying data, making operations fast and memory-efficient.

**Basic Rules of Broadcasting**

When performing operations on two arrays, NumPy compares their shapes element-wise, starting from the trailing dimensions (i.e., from the rightmost dimension) and applies these rules:

If the dimensions are equal, they are compatible.

If one of the dimensions is 1, it is stretched (broadcast) to match the other.

If the dimensions are different and neither is 1, broadcasting is not possible and an error is raised.

```python
# Example Compatible Shapes:

import numpy as np

a = np.array([1, 2, 3])      # Shape: (3,)
b = np.array([[10], [20]])   # Shape: (2,1)

# Broadcasting shapes: (2,1) and (3,) → Resulting shape: (2,3)
result = b + a
print(result)
```

**4. Explain the use of the groupby() method in Pandas?**

**Answer:** The groupby() method in Pandas is used to split data into groups based on one or more columns, and then apply a function (like aggregation, transformation, or filtering) to each group independently. It’s one of the most powerful and commonly used features in Pandas for data analysis.

The Concept: Split → Apply → Combine

Split the data into groups based on some criteria (e.g., values in a column).

Apply a function to each group (e.g., sum, mean, count).

Combine the results into a new DataFrame or Series.

```python
# Example:
Syntax:
df.groupby('column_name').function()
```

```python
import pandas as pd

# Sample data
data = {
    'Department': ['HR', 'IT', 'HR', 'Finance', 'IT'],
    'Employee': ['Tharn', 'prakesh', 'sreya', 'sarika', 'sunjana'],
    'Salary': [50000, 60000, 52000, 58000, 62000]
}

df = pd.DataFrame(data)

# Group by Department and calculate average salary
result = df.groupby('Department')['Salary'].mean()
print(result)
```

```python
# Multiple Aggregations Example:

df.groupby('Department')['Salary'].agg(['mean', 'max', 'min'])
```

**7.  What is a heatmap, and when should it be used?**

**Answer:** A heatmap is a data visualization technique that uses color gradients to represent the magnitude or intensity of values in a matrix or table. It helps in visually identifying patterns, correlations, or anomalies within the data.

** Key Characteristics of a Heatmap:**

Uses color to encode data values.

Typically used to show values in a 2D matrix format.

Each cell's color reflects the value at the intersection of a row and a column.

```python
#  Example in Python using Seaborn:
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample correlation matrix
data = pd.DataFrame({
    'Math': [80, 90, 70, 65],
    'Science': [88, 95, 60, 70],
    'English': [78, 85, 75, 80]
})

# Compute correlation matrix
corr = data.corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Subject Score Correlation Heatmap")
plt.show()
```

**17. Explain the difference between apply() and map() in Pandas**

**Answer:** Both apply() and map() are used to manipulate or transform data in a Pandas DataFrame or Series, but they serve different purposes and work in different contexts.

**1. map()**

Used only on Pandas Series (typically one column).

Applies a function, dictionary, or series to each element in the Series.

Great for element-wise transformations or value replacements.

```python
# Example:

import pandas as pd

# Sample Series
s = pd.Series([1, 2, 3, 4])

# Using map to square each value
s.map(lambda x: x ** 2)
```

```python
s = pd.Series(['cat', 'dog', 'bird'])
s.map({'cat': '🐱', 'dog': '🐶', 'bird': '🐦'})
```

**2. apply()**

Works on both Series and DataFrames.

For a Series, behaves like map() (element-wise).

For a DataFrame, it can apply a function row-wise or column-wise using axis parameter.

```python
# Example (on Series):
s = pd.Series([1, 2, 3])
s.apply(lambda x: x ** 2)
```

```python
# Example:

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Apply sum row-wise
df.apply(sum, axis=1)
```

**19.  How does Pandas simplify time series analysis?**

**Answer:** Pandas is one of the most powerful Python libraries for time series data. It simplifies time series analysis through easy-to-use datetime indexing, resampling, frequency conversion, and built-in time-aware operations.

**21.  Why is NumPy’s array slicing faster than Python’s list slicing?**

**Answer:** NumPy's array slicing is significantly faster and more memory-efficient than Python list slicing due to the way NumPy handles memory and data types.

**Conclusion:**

NumPy slicing is faster because it avoids memory copying, uses efficient memory layout, and leverages low-level optimizations. This is a major reason NumPy is preferred in data science, machine learning, and numerical computing.

```python
# Example: Timing the Difference

import numpy as np
import time

# NumPy
a = np.arange(1000000)
start = time.time()
b = a[100:999000]
end = time.time()
print("NumPy slicing time:", end - start)

# Python list
lst = list(range(1000000))
start = time.time()
lst2 = lst[100:999000]
end = time.time()
print("List slicing time:", end - start)
```
